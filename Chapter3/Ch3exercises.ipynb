# Task
You're building a movie search tool using the OMDb API (Open Movie Database). It's free and simple — perfect for practice. Your API key: trilogy (a public demo key) Base URL: https://www.omdbapi.com/

```python
import requests

# Build reusable wrapper function
def make_request(payload=None):
    return requests.get(
        'http://www.omdbapi.com/'
        params={'apikey': 'trilogy', **(payload or {})}
    )


# build a search_movies function which takes one argument
def search_movies(title):
    """search for movie by title."""
    response = make_request(payload={"s": title})
    return response.json()

# build a function to get the full details of one specific movie 
def get_movie_details(imdb_id):
    """use imdb id get full details of movie."""
    response = make_request(payload={"i": imdb_id})
    return response.json()

# build a function to search for the movie by title and filter by year
def get_movies_by_year(title, year):
    """use paramteers s and y to get movie title and filter by year."""
    response = make_request(payload={"s": title, "y": year})
    return response.json()


# test main function
if __name__ == "__main__":

    # search for movie batman
    search = search_movies("Batman")
    print("Selected movie:", search)

    # get movie details
    movie_details = get_movie_details("tt0372784")
    print("Movie details:", movie_details)

    # search for spiderman 2002
    get_movie = get_movies_by_year("Spider-Man", "2002")
    print("Movie picked:", get_movie)
```
